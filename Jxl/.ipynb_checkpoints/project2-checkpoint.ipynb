Project background: In recent years, with the rapid development of Internet of Vehicles technology, people can carry out statistical analysis of drivers' driving behaviors-including driving habits analysis, driver driving behavior portraits, and can also provide early warning of drivers' bad driving behaviors.

The data used comes from a website, contains 8 vehicle information, divided into 754 segments. During driving, data is recorded every second or every 15 seconds. The data contains 14 indicators (all continuous), divided into 5 dimensions: time dimension, speed dimension, mileage dimension, mechanical dimension, driving stability.

- Time dimension: driving time, fatigue driving time, morning and evening rush hours, night drive
- Average speed, maximum speed, speed standard deviation, congestion, high speed drive
- Mileage dimension: mileage
- Mechanical performance: average engine speed, maximum engine speed, standard deviation of engine speed
- Driving stability: stability

## Data preprocessing

```python
# import package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data
df2 = pd.read_csv("data2.csv")

# Sort vehicle data by category
df2_car1 = df2[df2['NUM']==1]
df2_car2 = df2[df2['NUM']==2]
df2_car3 = df2[df2['NUM']==3]
df2_car4 = df2[df2['NUM']==4]
df2_car5 = df2[df2['NUM']==5]
df2_car6 = df2[df2['NUM']==6]
df2_car7 = df2[df2['NUM']==7]
df2_car8 = df2[df2['NUM']==8]

# Data preprocessing
del df2_car1['NUM']
del df2_car2['NUM']
del df2_car3['NUM']
del df2_car4['NUM']
del df2_car5['NUM']
del df2_car6['NUM']
del df2_car7['NUM']
del df2_car8['NUM']
```

## Data Analysis of Car 1

```python
# Descriptive statistics
df2_car1.hist(figsize=(15,10))
df2_car1.describe()
```

```python
# Fatigue driving time, morning and evening peak hours, and late night travel time ratio
df2_car1.sum()
ylabel = [df2_car1.sum()["NIGHT"],df2_car1.sum()["RUSHOUR"],df2_car1.sum()["TIRED"]]
xlabel = ["Night" , "Rushhour" , "Tired"]
plt.bar(xlabel , ylabel)

for a,b in zip(xlabel,ylabel):
    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=7)

plt.show()
```

## Comparison between 8 car drivers

```python
# atigue driving time, morning and evening peak hours, and late night travel time ratio fpr 8 drivers

ylabel1 = [df2_car1.sum()["NIGHT"],df2_car1.sum()["RUSHOUR"],df2_car1.sum()["TIRED"]]
ylabel2 = [df2_car2.sum()["NIGHT"],df2_car2.sum()["RUSHOUR"],df2_car2.sum()["TIRED"]]
ylabel3 = [df2_car3.sum()["NIGHT"],df2_car3.sum()["RUSHOUR"],df2_car3.sum()["TIRED"]]
ylabel4 = [df2_car4.sum()["NIGHT"],df2_car4.sum()["RUSHOUR"],df2_car4.sum()["TIRED"]]
ylabel5 = [df2_car5.sum()["NIGHT"],df2_car5.sum()["RUSHOUR"],df2_car5.sum()["TIRED"]]
ylabel6 = [df2_car6.sum()["NIGHT"],df2_car6.sum()["RUSHOUR"],df2_car6.sum()["TIRED"]]
ylabel7 = [df2_car7.sum()["NIGHT"],df2_car7.sum()["RUSHOUR"],df2_car7.sum()["TIRED"]]
ylabel8 = [df2_car8.sum()["NIGHT"],df2_car8.sum()["RUSHOUR"],df2_car8.sum()["TIRED"]]
xlabel = ["Night" , "Rushhour" , "Tired"]

xlabel_num = np.arange(3)

width = 0.05
bar1= plt.bar(xlabel_num, ylabel1,width=width)
bar2= plt.bar(xlabel_num+width, ylabel2,width=width)
bar3= plt.bar(xlabel_num+width*2, ylabel3,width=width)
bar4= plt.bar(xlabel_num+width*3, ylabel4,width=width)
bar5= plt.bar(xlabel_num+width*4, ylabel5,width=width)
bar6= plt.bar(xlabel_num+width*5, ylabel6,width=width)
bar7= plt.bar(xlabel_num+width*6, ylabel7,width=width)
bar8= plt.bar(xlabel_num+width*7, ylabel8,width=width)

plt.xticks(xlabel_num+width*2 , xlabel)
plt.legend(handles=[bar1,bar2,bar3,bar4,bar5,bar6,bar7,bar8],labels= ["1","2","3","4","5","6","7","8"])

plt.show()
```

Judging from the above picture, car 2 and car 4 have a higher proportion of night driving. 
The rushhours of 8 vehicles have been significantly higher than other time. 4, 6, 7, 8 car owners may work in the city center, so the rushhours time is relatively high. 
Finally, the fatigue driving time of 8 vehicles is almost the same, less than 10h.
