```python
# import package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 时间加速度图
df1 = pd.read_csv("data1.csv") 
num = len(df1)

# plt.figure(figsize=(50, 20))

# 添加1s时间点
df1["TIME_1s"]=pd.date_range(start='2016-11-11 19:50', periods=num,freq="S")
df1.set_index("TIME_1s")
df1.plot(x="TIME_1s",y="LONGACC",title='TIME and LONGACC')

# my_x_ticks =  设置x轴标签
# plt.xticks(my_x_ticks)
plt.xlabel("Time")
plt.ylabel("Accelerate")


df1.plot(x="TIME_1s",y="SPEED",title='TIME_1s and SPEED')

df1.plot(x="TIME_1s",y="CONSUM",title='TIME_1s and CONSUM')

df1.plot(x="TIME_1s",y="WHEEL",title='TIME_1s and WHEEL')

df1.plot(x="TIME_1s",y="CUMDIS",title='TIME_1s and CUMDIS')
```

```python
# 前后时刻的图
# diff_Acc = df1["LONGACC"][1:len(df1)]-df1["LONGACC"][0:(len(df1)-1)]

# 后一时刻的速度
speed_1 =  df1["SPEED"][1:num]
# 前后时刻的速度比
plt.scatter(df1["SPEED"][0:(num-1)],speed_1)
plt.xlabel("V_t")
plt.ylabel("V_t+1")

# diff_Speed = df1["SPEED"].diff()
# diff_Speed
# df1["diff_Speed"]= diff_Speed
# df1
```
