# Data preprocessing

```python
import pandas as pd
from math import radians, cos, sin, asin, sqrt,ceil
import numpy as np
import geohash
import matplotlib.pyplot as plt
import requests
import folium
from folium import plugins



data_bike = pd.read_csv("2016_mobike_shanghai.csv")
```

```python
# load data
data_bike.shape

data_bike.info()
```

```python
# convert time format
data_bike['start_time'] = pd.to_datetime(data_bike['start_time'])
data_bike['end_time'] = pd.to_datetime(data_bike['end_time'])
```

- Calculate the riding time, the unit is min.

```python
# data analyze

# Riding time the unit is min
duration = data_bike.end_time - data_bike.start_time
data_bike["duration"] = (data_bike.end_time - data_bike.start_time).dt.seconds/60
```

- Calculate the distance of the ride by the latitude and longitude of the starting point and the end of the ride, the unit is kilometers

```python
# Calculate the distance of the ride by the latitude and longitude 
# of the starting point and the end of the ride, the unit is kilometers

def bike_distance(item):
    longitude_start, latitude_start, longitude_end, latitude_end = map(radians, [item["start_location_x"], item["start_location_y"], item["end_location_x"], item["end_location_y"]]) # 经纬度转换成弧度
    dist_longitude = longitude_start - longitude_end
    dist_latitude = latitude_start - latitude_end
    
    dist = sin(dist_latitude/2)**2 + cos(latitude_start) * cos(latitude_end) * sin(dist_longitude/2)**2
    
    distance = round(2 * asin(sqrt(dist)) * 6371,3)  
    return distance

data_bike["dist"] = data_bike.apply(bike_distance,axis=1)
```

- Obtain the route through the trail of each bike 

```python
# Obtain the route through the trail of each bike 

def geo_Length(item):
    track_list = item["track"].split("#")  
#     print(track_list)
    
    part_Length_item = {}
    global Length 
    Length = 0
    
    for i in range(len(track_list)-1):
        
        start_location = track_list[i].split(",")
        end_location = track_list[i+1].split(",")
        
        part_Length_item["start_location_x"],part_Length_item["start_location_y"] = float(start_location[0]),float(start_location[1])
        part_Length_item["end_location_x"],part_Length_item["end_location_y"] = float(end_location[0]),float(end_location[1])
        
        # distance
        part_length_each = bike_distance(part_Length_item)
        Length = part_length_each + Length
        
    return Length


data_bike["Length"] = data_bike.apply(geo_Length,axis=1)

data_bike["Length"] 
```

- Get the day of the week corresponding to the date, like weekday and weekend

```python
# weekday and clear time

# weekday
data_bike['weekday'] = data_bike.start_time.apply(lambda x: x.isoweekday())
# clear time  the unit is h
data_bike['daytime'] = data_bike.start_time.apply(lambda x: x.utctimetuple().tm_hour)
```

- Get the number of bike use .

```python
# numbers of bike use
bike_use_times = data_bike.groupby("bikeid")["orderid"].count()
```

- Calculate the price per order ,we set 1 ￥ per 20min

```python
# cost per ride
# Calculate the price per order  
# we set 1 ￥ per 20min

data_bike['cost'] = data_bike.duration.apply(lambda x: ceil(x/20))
```

## Analysis of parking spots

```python
def district_diff(item):
    item["start_location_x"] = round(item["start_location_x"],2)
    item["start_location_y"] = round(item["start_location_y"],2)
    item["end_location_x"] = round(item["end_location_x"],2)
    item["end_location_y"] = round(item["end_location_y"],2)
    
    return item

data_bike = data_bike.apply(district_diff,axis = 1)
```

```python
data_bike["date"] = data_bike["start_time"].dt.day

max_data_ridenum = data_bike[data_bike["date"]==data_bike["date"].value_counts().index[0]]

earlist = max_data_ridenum.groupby(["bikeid"]).apply(lambda t: t[t.start_time==t.start_time.min()])
last = max_data_ridenum.groupby(["bikeid"]).apply(lambda t: t[t.start_time==t.start_time.max()])

district_out_num = earlist.groupby(["start_location_x","start_location_y"]).count()["bikeid"]
district_back_num = last.groupby(["end_location_x","end_location_y"]).count()["bikeid"]

district_out_back = pd.DataFrame({"out":district_out_num,"back":district_back_num})
district_out_back.fillna(0,inplace=True)

# the number gap of bike out and back 
district_out_back["num_difference"] = district_out_back.apply(lambda t: t["back"]-t["out"],axis = 1)
```

```python
lack_bike_district = district_out_back[(district_out_back["num_difference"]) <= -5].sort_values(by=["num_difference"])

free_bike_district = district_out_back[(district_out_back["num_difference"]) >= 5].sort_values(by=["num_difference"])
```

We use the open API of the AutoNavi(Gaode) platform to obtain the specific address corresponding to the latitude and longitude

```python
# get the clear location
def API_location(location):
    
    location_str=str(location[0])+','+str(location[1])
    parameters = {'output': 'json', 'location': location_str, 'key': '2fbbe1979c49ea25e83e821988513d97','extensions':'all'}
    base = 'https://restapi.amap.com/v3/geocode/regeo'
    
    response = requests.get(base, parameters)
    answer = response.json()["regeocode"]["formatted_address"]
    return answer
```

```python
lack_location = []

for i in lack_bike_district.index:
    item_result = API_location(i)
    lack_location.append(item_result)
    
lack_bike_district["location"] = lack_location
```

```python
free_location = []

for i in free_bike_district.index:
    item_result = API_location(i)
    free_location.append(item_result)
    
free_bike_district["location"] = free_location
```
